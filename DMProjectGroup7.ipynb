## Problem 1:
The problem in existing image classification methods is that they employ fixed weight fusion strategies where images with different semantics are forced to share the same weights, which results in poorer and less sharper edge detection. This sometimes results in the predicted edge-maps losing some edges in the images.

This problem is addressed in both papers by a sub-derivative of deep learning models in Convolutional Neural Networks. They use Dynamic feature fusion to differentiate between low level and high level features in images and use this to fine tune the edge detection.

This problem is interesting because instead of relying on classic and low level methods, we can unleash the power of DL in edge detection to get more accurate results. This can also enhance the practical application of edge detection in fields such as Medical Image Analysis and even for Cityscape planning images.



## Problem 2:

### 1)
**1)** The first paper uses Dynamic Feature Fusion method that assigns different fusion weights adaptively to distinct input images and locations. This is to generate a fused edge map adaptive to the specific content of each image. This is done by using a location-adaptive weight learner to infer proper fusion weights over multi-level features for each location of the feature map, conditioned on the specific input.\
The proposed DFF model consists of two main components: 1) A feature extractor with a normalizer which scales the multi-level features to the same magnitude 2)An adaptive weight fusion module which dynamically generates location-specific fusion weights and uses it to combine low level and high level response maps to produce the final fused output.\
Instead of using CaseNet which uses a fixed weight system, ResNet along with DFF is used for sharper edge detection. The dataset mainly used are Cityscape images and SBD with varying scale factors.\
Given the activation of the multi-level side outputs Aside of size H X W, we aim to obtain a fused output Afuse = f(Aside) by aggregating multi-level responses.\
After fusion, the adaptive weight learner is formulated as

 Afuse = f(Aside; W)
 
Location-adaptive weight learner is better than Location-invariant weight learner and can be used to overcome fixed weight fusion. Ablation experiments are conducted on Cityscapes to investigate the effectiveness of each newly proposed module. Normalizer scales are used to tone down the fusion models and add clarity and sharpness to the edge images as result.


**2)** The second paper uses DexiNed which is also based on Convolutional Neural Networks(CNN) which is a part of Deep Learning. It has two sub networks: Dense Extreme Inception(DEXI) Network along with Upsampling block(UB). Dexi provides feature maps which are fed into UB which in turn produces a thin-edge map as its output. There are 6 main blocks in this method, each having its own convolution and UB. After the 3rd block, the output of each sub-block is averaged with edge-connection so as to not lose the important edge features. DexiNed uses standard convolutions. UB consists of a convolutional and deconvolutional block to better retain sharper edges.The filters are also reduced from 728 to 512 in block 4 and block 5.\
The two major variants of Dexined are Dexined-A(Averaged) and Dexined-F(Fused). Fused is the concatenation and fusion of previous outputs while averaged is the average of all previous predictions. Both perform well depending on the dataset obtained to give better edge detection, while even on poor detection, the results are far better.\
The dataset used for edge detection here is MDBD( Multicue Dataset for Boundary Detection) while for boundary detection, CID, BSDS, NYUD and PASCAL datasets are used.\
DexiNed could be summarized as a regression function ð, that is, Yˆ = ð(X, Y ), where X is an input image, Y is its respective ground truth, and Yˆ is a set of predicted edge maps. BIPED(Barcelona Images for Perceptual Edge Detection) is used to train data just once, and its result can outperform classic and latest models.

One method uses location adaptive weight learner while other method implements fused and average methods to get better edge detection.\
Edge and boundary detection can both be done based on the dataset provided in the Dexined method while edge detection can only be done in the Semantic edge detection method.\

## Problem 9

The major pitfall would be the training time taken for the models, complexity of the models and quality of datasets used for the models. Models with deeper network generate better multi-scale features but also are harder to train, involve more parameters and take up lot of resources.

Deep Learning models strongly depend on the quality of the datasets for training and evaluation. Models trained on datasets with better annotated images will give better results. Making changes in the architecture of the models to make the network lighter can reduce complexity. Replacing the blocks in the architecture with lower number of more efficient blocks containing multiple convolution layers can give better multi-scale features and in turn generate better edges.

Canny Edge detection along with Fuzzy Inference: Canny edge detection filter is a multi stage edge detector which uses a derivative of gaussian filter and aids in obtaining sharp and smooth edges. Advanced Canny filters can implement dynamic fusion which assist us to provide better edge detection and when coupled with fuzzy inference, it reduces the blur of edges and reduces noise. This also has higher accuracy and can accurately localize on the center of the edge. The detected edges can be thinned down to 1-pixel level, aiding us to obtain a sharper edge. Fuzzy image processing is a powerful tool to fuzzify the images and then map a range of values to detect  floating matrices to detect the edge. This method is superimposed on canny to get the desired results and since stages involved are fairly minimal, training time is less and hence, it can obtain high accuracy results faster.
